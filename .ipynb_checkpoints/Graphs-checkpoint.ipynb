```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_k05 = np.loadtxt("mae_s_k05.txt")
x_k10 = np.loadtxt("mae_s_k10.txt")
x_k20 = np.loadtxt("mae_s_k20.txt")
x_k40 = np.loadtxt("mae_s_k40.txt")
```

```python
plt.figure(figsize=(12,6))
plt.xlabel("Iterations/10")
plt.ylabel("Mean Absolute error")
plt.plot(x_k05,'red',label="K=5")
plt.plot(x_k10,'blue',label="K=10")
plt.plot(x_k20,'green',label="K=20")
plt.plot(x_k40,'black',label="K=40")
plt.legend(loc="upper right")
plt.grid()
plt.savefig('x_matrix.png')
plt.show()
```

```python
y_k05 = np.loadtxt("mae_s_y_k05.txt")
y_k10 = np.loadtxt("mae_s_y_k10.txt")
y_k20 = np.loadtxt("mae_s_y_k20.txt")
plt.figure(figsize=(12,6))
plt.xlabel("Iterations/10")
plt.ylabel("Mean Absolute error")
plt.plot(y_k05,'red',label="K=5")
plt.plot(y_k10,'blue',label="K=10")
plt.plot(y_k20,'green',label="K=20")
plt.legend(loc="upper right")
plt.grid()
plt.savefig('y_matrix.png')
plt.show()
```

```python
d_k05 = np.loadtxt("mae_d_k05.txt")
d_k10 = np.loadtxt("mae_d_k10.txt")
plt.figure(figsize=(12,6))
plt.xlabel("Iterations/10")
plt.ylabel("Mean Absolute error")
plt.plot(y_k05+x_k05,'red',label="Sum of MAEs for K=5")
plt.plot(y_k10+x_k10,'blue',label="Sum of MAEs for K=10")
plt.plot(d_k05,'green',label="Extended HPF MAE for K=05")
plt.plot(d_k10,'black',label="Extended HPF MAE for K=10")
plt.legend(loc="upper right")
plt.grid()
plt.savefig('dual_matrix.png')
plt.show()
```
