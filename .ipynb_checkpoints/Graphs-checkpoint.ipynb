```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_k05 = np.loadtxt("mae_s_k05.txt")
x_k10 = np.loadtxt("mae_s_k10.txt")
x_k20 = np.loadtxt("mae_s_k20.txt")
x_k40 = np.loadtxt("mae_s_k40.txt")
```

```python
plt.figure(figsize=(20,10))
plt.plot(x_k05,'red',label="K=5")
plt.plot(x_k10,'blue',label="K=10")
plt.plot(x_k20,'green',label="K=20")
plt.plot(x_k40,'black',label="K=40")
plt.legend(loc="upper right")
plt.grid()
plt.show()
```
